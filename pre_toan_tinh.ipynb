```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
file_names = ["ban_mau_12_11_21.csv",
             "ban_mau_12_13_21.csv",
             "ban_mau_12_14_21.csv",
             "ban_mau_12_15_21.csv",
             "ban_mau_12_17_21.csv",
             "ban_mau_12_19_21.csv",
             "ban_mau_12_20_21.csv",
             "ban_mau_12_21_21.csv",
             "ban_mau_12_22_21.csv",
             "ban_mau_12_23_21.csv"]
```

```python
dfs = []
for name in file_names:
    dfs.append(pd.read_csv(name))
```

```python
pd.merge(dfs[1],dfs[2], on=["huyen", "xa", "thon"],suffixes=('_left', '_right'))
```

```python
from functools import reduce
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on=["huyen", "xa", "thon"]), dfs)
```

```python
df_final
```

```python
a = len(df_final.columns[3:])
```

```python
list(range(a))
```

```python
columns = ['huyen', 'xa', 'thon'] + list(range(a))
```

```python
df_final.columns = columns
```

```python
data = df_final[df_final.columns[3:]].to_numpy()
```

```python
for i in data:
    print(i,'hello')
```

```python
x = []
y = []
for i in data:
    for j in range(4):
        x.append(i[j:j+5])
        y.append(i[j+5])
```

```python
x
```

```python
x_train = []
for x_tr in x:
    x_dem = [0, 0, 0, 0]
    for (i,x_tri) in enumerate(x_tr):
        x_dem[x_tri-1]+=(1+i)
    x_train.append(x_dem) 
```

```python
y
```

```python
x_train
```

```python
from __future__ import print_function
from sklearn.naive_bayes import MultinomialNB
import numpy as np 
```

```python
clf = MultinomialNB()
```

```python
clf.fit(x_train, y)
```
